```python
document=[]
print(movie_reviews.categories())
for category in movie_reviews.categories():
    for id in movie_reviews.fileids(category):
        document.append((movie_reviews.words(id),category))
document
```

```python
#shuffling the data inorder to split
import random
random.shuffle(document)
document[0:5]
```

```python
from nltk.corpus import stopwords
```

```python
import string
stops=stopwords.words('english')
punctuations=string.punctuation
stops+=punctuations
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemma=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_pos(tag):
    if tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
```

```python
def clean_review(words):
    output=[]
    for w in words:
        if w.lower() not in stops:
            k=pos_tag([w])
            clean_words=lemma.lemmatize(w,pos=get_pos(k[0][1]))
            output.append(clean_words.lower())
    return output
    
```

```python
cleaned_document= [(clean_review(doc),category) for doc,category in document]
```

```python
cleaned_document[0]
```

```python
training_data=cleaned_document[0:1500]
testing_data=cleaned_document[1500:]
```

```python
all_words=[]
for j in training_data:
    all_words+=j[0]
all_words
```

```python
import nltk
```

```python
#it is neither list nor dictionary(tuple) .But it is frequency distribution object
freq=nltk.FreqDist(all_words)
```

```python
common=freq.most_common(3000)
feature=[i[0] for i in common]
feature
```

```python
def get_feature_dict(words):
    current_feature={}
    words_set=set(words)
    for f in feature:
        current_feature[f]=f in words_set
    return current_feature
```

```python
get_feature_dict(training_data[0][0])
```

```python
training_data_points=[(get_feature_dict(doc),category) for doc,category in training_data]
```

```python
testing_data_points=[(get_feature_dict(doc),category) for doc,category in testing_data]
```

```python
training_data_points[0]
```

```python
#we created the data in above format becuase nltk classifier wants our data to be in that format
from nltk import NaiveBayesClassifier
```

```python
classifier=NaiveBayesClassifier.train(training_data_points)
```

```python
#This score is basically on training-data because in our testing_points we took train_data ,see above.
nltk.classify.accuracy(classifier,testing_data_points)
```

```python
classifier.show_most_informative_features(15)
```

# Using nltk only for cleaning data and sklearn classifiers to predict the result

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
categories=[category for doc,category in cleaned_document]
```

```python
text_document=[" ".join(doc) for doc,category in cleaned_document]
```

```python
text_document[0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(text_document,categories)
```

```python
# Use of N-Grams in your feature space may not necessarily yield any significant improvement.
count=CountVectorizer(max_features=2000,ngram_range=(1,2),max_df=0.8,min_df=0.1)
```

```python
x_train_features=count.fit_transform(x_train)
x_train_features.todense()
```

```python
count.get_feature_names()
```

```python
x_test_features=count.transform(x_test)
x_test_features.todense()
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC()
svc.fit(x_train_features,y_train)
```

```python
y_pred=svc.predict(x_test_features)
svc.score(x_test_features,y_test)
```
